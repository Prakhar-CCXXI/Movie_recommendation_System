```python
import ast
df['genres'] = df['genres'].apply(lambda x :" ".join([i['name'] for i in ast.literal_eval(x)]))
```

 What ast.literal_eval does it evaluates that string and safely converts it into an actual, usable Python list of dictionaries.
 
 [i['name'] for i in ...] It loops through each dictionary (i) inside your newly converted list and grabs only the value attached to the key 'name'.
 
 " ".join(...)This takes your clean list of genre strings and glues them together into a single string, using a space " " as the separator.
 
 .apply(lambda x: ...) function tells pandas to run this entire extraction process on every single row (x) down the genres column, overwriting the old messy text with the new clean text.

# NLP Text Processing

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

nltk -> The parent library that provides all human-language processing tools.
stopwords -> Drops useless filler words like "the", "and", and "is".
WordNetLemmatizer -> Flattens words to their base dictionary roots like converting "running" to "run".
re ->  Strips out annoying punctuation, symbols, and numbers using pattern matching.

```python
nltk.download('stopwords')
nltk.download('wordnet')
# nltk.download('omw-1.4')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    #lowercase
    text = str(text).lower()
    #removed the punctuations
    text = re.sub('[^\w\s]|\d|_', ' ', text)
    # Split text into individual words
    words = text.split()
    
    # Remove stop words AND lemmatise the remaining words
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

    # Glue the cleaned words back into a single clean string
    return " ".join(words)
```

```python
df['tags']= df['tags'].apply(preprocess_text)
```

# Text Vectorization

```python
df = df.reset_index(drop=True)
```

```python
indices = pd.Series(df.index,index=df['title']).drop_duplicates()
```

```python
indices
```

```python
## TFIDF Vectorization evaluates how important a word is to a document within a collection or corpus [1]. It is calculated by multiplying two metrics
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000,ngram_range=(1, 2),stop_words = 'english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
    if title not in indices:
        return ['Movie not found']


    idx = indices[title]
    if isinstance(idx, pd.Series):
        idx = idx.iloc[0]
        
    sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]

    return df['title'].iloc[similar_idx].tolist()
    # for i in similar_idx:
    #     print(df.iloc[i]['title'], sim_score[i])
```

```python
recommend('Avatar',5)
```

```python
import pickle

pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))

df.to_pickle('df.pkl')

pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
